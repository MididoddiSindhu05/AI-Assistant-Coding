#Task - 02(Incorrect condition in an If Statement)

```python
def check_number(n):
    if n == 10:
        return "Ten"
    else:
        return "Not Ten"
assert check_number(10) == "Ten"
assert check_number(5) == "Not Ten"
assert check_number(0) == "Not Ten"
```

#Task - 03(Runtime Error – File Not Found)

```python
def read_file(filename):
    try:
        with open(filename, 'r') as f:
            return f.read()
    except FileNotFoundError:
        return f"Error: The file '{filename}' was not found."

print(read_file("nonexistent.txt"))
with open("test.txt", "w") as f:
    f.write("Hello File")

assert read_file("test.txt") == "Hello File"

assert read_file("missing.txt") == "Error: The file 'missing.txt' was not found."

assert read_file("/invalid/path/file.txt") == "Error: The file '/invalid/path/file.txt' was not found."
```

#Task - 04(Calling a Non-Existent Method)

```python
class Car:
    def start(self):
        return "Car started"

    def drive(self):
        return "Car is driving"

my_car = Car()
print(my_car.start())

car = Car()

assert car.start() == "Car started"
assert car.drive() == "Car is driving"
assert isinstance(car.drive(), str)
```

#Task - 05(TypeError – Mixing Strings and Integers in Addition)

```python
def add_five(value):
    return int(value) + 5
print(add_five("10"))
assert add_five("10") == 15
assert add_five(5) == 10
assert add_five("0") == 5
```
